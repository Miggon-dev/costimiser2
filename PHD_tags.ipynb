````python
def get_line(path, line_no, encoding="utf-8"):
    with open(path, "r", encoding=encoding, errors="strict") as f:
        for i, line in enumerate(f, start=1):
            if i == line_no:
                return line.rstrip("\n")
    raise ValueError(f"Line {line_no} not found")
 
# Known from your diagnosis
desc_lines = [get_line(path, i) for i in [1, 2, 3, 4]]  # wrapped descriptions
units_line = get_line(path, 5)
tags_line  = get_line(path, 6)
 
# Split units/tags (these are full-width already)
units = units_line.split("\t")
tags  = tags_line.split("\t")
 
# Fix first column name/unit (blank -> Time)
tags[0] = "Time"
units[0] = "datetime"
 
expected_cols = len(tags)  # should be 2275
 
# --- Reconstruct descriptions by concatenating pieces from lines 1-4 ---
desc = []
for k, line in enumerate(desc_lines):
    parts = line.split("\t")
    if k == 0:
        # line 1 includes "Time" then first chunk of descriptions
        # keep same length as header chunk
        desc.extend(parts)
    else:
        # lines 2-4 start with a blank first field -> drop it
        if len(parts) > 0 and parts[0].strip() == "":
            parts = parts[1:]
        desc.extend(parts)
 
# Ensure descriptions align to expected column count
# (pad/truncate defensively in case of small off-by-few issues)
if len(desc) < expected_cols:
    desc = desc + [""] * (expected_cols - len(desc))
else:
    desc = desc[:expected_cols]
 
# Make sure first description is "Time" (optional)
if desc[0].strip() == "":
    desc[0] = "Time"
 
meta = pd.DataFrame({"tag": tags, "unit": units, "description": desc})
 
print("meta shape:", meta.shape)
print(meta.head(10))
````

````python
import re
import unicodedata
import pandas as pd
 
# --- compiled regexes (fast) ---
 
# tag-like substrings such as 2-033-FI100-D2 or 2-033-FC100
TAG_LIKE = re.compile(r"\b\d+-\d+-[A-Z0-9]+(?:-[A-Z0-9]+)?\b")
 
# complete bracket blocks: [ ... ]
FULL_BRACKET = re.compile(r"\[[^\]]*\]")
 
# dangling bracket interval at end (no closing ])
DANGLING_INTERVAL = re.compile(
    r"\[\s*[-+]?\d+(?:[.,]\d+)?\s*\.\.\s*[-+]?\d+(?:[.,]\d+)?[^]]*$"
)
 
# any remaining lone '['
LONE_OPEN_BRACKET = re.compile(r"\[")
 
# suffix ~^n at end
SUFFIX = re.compile(r"\s*~\^(\d+)\s*$")
 
# invisible characters explicitly
ZW_BOM = re.compile(r"[\u200b\u200c\u200d\u2060\ufeff]")
 
# unit mentions (add more as needed)
UNIT_PATTERNS = [
    r"\(\s*%+\s*\)",
    r"\(\s*vol\s*%\s*\)",
    r"\(\s*vol%\s*\)",
 
    r"\b(in)\s+kwh\b",
    r"\b(in)\s+mwh\b",
    r"\b(in)\s+g\s*/\s*m(?:\^?2|²)\b",
 
    r"\bt\s*/\s*hr\b",
    r"\bl\s*/\s*h\b",
    r"\bl\s*/\s*min\b",
    r"\bm\s*/\s*min\b",
    r"\bm(?:\^?2|²)\b",
    r"\bg\s*/\s*m(?:\^?2|²)\b",
    r"\bm(?:\^?3|³)\s*/\s*h\b",
 
    r"\bkwh\b",
    r"\bmwh\b",
 
    r"\b\d+(?:[.,]\d+)?\s*(?:-|–|\.\.)\s*\d+(?:[.,]\d+)?\s*bar\b",
]
UNITS_RE = re.compile("|".join(f"(?:{p})" for p in UNIT_PATTERNS), flags=re.IGNORECASE)
 
# logic notes to remove
LOGIC_RE = re.compile(
    r"(?:\(\s*0\s*=\s*false\s*\))"
    r"|(?:/\s*no\s*tear\s*=\s*-?1\s*\|\s*tear\s*=\s*0)",
    flags=re.IGNORECASE
)
 
def clean_meta_text(x: object) -> object:
    """Apply all cleaning steps to a single cell."""
    if pd.isna(x):
        return x
 
    s = str(x)
 
    # 1) remove zero-width + BOM explicitly
    s = ZW_BOM.sub("", s)
 
    # 2) remove other unicode control chars (category starts with 'C')
    s = "".join(ch for ch in s if unicodedata.category(ch)[0] != "C")
 
    # 3) ~^n -> (n+1)
    m = SUFFIX.search(s)
    if m:
        n = int(m.group(1))
        s = SUFFIX.sub(f" ({n+1})", s)
 
    # 4) remove tag-like substrings
    s = TAG_LIKE.sub("", s)
 
    # 5) remove proper bracket blocks and dangling intervals
    #s = FULL_BRACKET.sub("", s)
    #s = DANGLING_INTERVAL.sub("", s)
    #s = LONE_OPEN_BRACKET.sub("", s)
 
    # 6) remove logic notes
    #s = LOGIC_RE.sub("", s)
 
    # 7) remove unit mentions
    #s = UNITS_RE.sub("", s)
 
    # 8) normalize whitespace and tidy punctuation leftovers
    s = re.sub(r"\s+", " ", s).strip()
    s = re.sub(r"\s*:\s*$", "", s).strip()

       # 8) normalize whitespace
    s = re.sub(r"\s+", " ", s).strip()
 
    # Remove trailing colon
    s = re.sub(r"\s*:\s*$", "", s).strip()
 
    # Remove leading/trailing dots
    s = re.sub(r"^\.+\s*", "", s)   # dots at start
    s = re.sub(r"\s*\.+$", "", s)   # dots at end
 
    # Final trim
    s = s.strip()
 
    return s
 
def clean_meta_dataframe(meta: pd.DataFrame, columns: list[str]) -> pd.DataFrame:
    """Clean selected columns in meta in-place and return meta."""
    for c in columns:
        if c in meta.columns:
            meta[c] = meta[c].apply(clean_meta_text)
    return meta
 
# --- usage: choose the columns you want cleaned (German + English if present) ---
#cols_to_clean = ["machine_part", "description_text", "machine_part_en", "description_text_en"]
cols_to_clean = ["description"]
meta = clean_meta_dataframe(meta, cols_to_clean)
 
meta[cols_to_clean].head(15)
````

````python
import boto3
import json
import pandas as pd
from tqdm.auto import tqdm
import os

session = boto3.Session(
    #profile_name="DatafactoryAdministrator-636452271875",
    profile_name="DatafactoryDeveloper-781690932061",
    region_name="eu-west-1"
)
s3 = session.client("s3")

 
bedrock = session.client("bedrock-runtime", region_name="eu-west-1")
MODEL_ID = "arn:aws:bedrock:eu-west-1:781690932061:inference-profile/eu.anthropic.claude-sonnet-4-5-20250929-v1:0"

SYSTEM_PROMPT = """You are a professional industrial translator.
Translate German industrial variable names into precise technical English.
 
Rules:
- Preserve technical meaning.
- Do not expand plant codes or IDs.
- Do not explain.
- Output MUST be JSON Lines (JSONL): one JSON object per line.
- Each line format: {"src": "<original>", "tgt": "<translation>"}
- Output only JSONL lines, no markdown, no extra text.
"""
 
def _parse_jsonl(text: str) -> dict:
    if not text or not text.strip():
        raise ValueError("Empty model text")
 
    # remove code fences if any
    text = re.sub(r"^```(?:jsonl|json)?\s*", "", text.strip(), flags=re.IGNORECASE)
    text = re.sub(r"\s*```$", "", text)
 
    mapping = {}
    bad = 0
    for line in text.splitlines():
        line = line.strip()
        if not line:
            continue
        try:
            obj = json.loads(line)
            mapping[obj["src"]] = obj["tgt"]
        except Exception:
            bad += 1
 
    if not mapping:
        raise ValueError(f"No JSONL objects parsed. First 200 chars: {text[:200]!r}")
    return mapping
 
def translate_batch_bedrock_jsonl(strings, max_retries=2):
    user_prompt = "Translate these terms into English. Return JSONL only.\n"
    user_prompt += "\n".join(f"- {s}" for s in strings)
 
    body = {
        "anthropic_version": "bedrock-2023-05-31",
        "max_tokens": 3500,      # keep some headroom
        "temperature": 0,
        "system": SYSTEM_PROMPT,
        "messages": [{"role": "user", "content": user_prompt}],
    }
 
    last_err = None
    for attempt in range(max_retries + 1):
        resp = bedrock.invoke_model(modelId=MODEL_ID, body=json.dumps(body))
        payload = json.loads(resp["body"].read())
        text = payload.get("content", [{}])[0].get("text", "")
 
        try:
            out = _parse_jsonl(text)
            # Fill missing with original
            for s in strings:
                out.setdefault(s, s)
            return out
        except Exception as e:
            last_err = e
            body["messages"].append({
                "role": "user",
                "content": "Your output was not valid JSONL. Output ONLY JSONL lines, one per input bullet."
            })
            time.sleep(0.5)
 
    raise RuntimeError(f"Failed JSONL. Last error: {last_err}")
 
def translate_unique_with_bedrock(values, batch_size=50, min_batch_size=5):
    values = pd.Series(values).dropna().astype(str)
    values = values[values.str.strip() != ""]
    unique_vals = values.unique().tolist()
 
    mapping = {}
    i = 0
    pbar = tqdm(total=len(unique_vals), desc="Translating (Bedrock)", unit="str")
 
    while i < len(unique_vals):
        bs = min(batch_size, len(unique_vals) - i)
        batch = unique_vals[i:i+bs]
 
        try:
            out = translate_batch_bedrock_jsonl(batch)
            mapping.update(out)
            i += bs
            pbar.update(bs)
        except RuntimeError:
            # If it fails, reduce batch size and retry this same chunk
            if bs <= min_batch_size:
                # last resort: keep originals to avoid infinite loop
                for s in batch:
                    mapping.setdefault(s, s)
                i += bs
                pbar.update(bs)
            else:
                batch_size = max(min_batch_size, bs // 2)
 
    pbar.close()
    return mapping
 
# Example usage
desc_map = translate_unique_with_bedrock(meta["description"].unique(), batch_size=80)
 
meta["description_en"] = meta["description"].map(desc_map)
````
